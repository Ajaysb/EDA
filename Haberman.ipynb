```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
#Our dataset didn't had column names
header = ['age','op_yr','aux_nodes','surv']
df = pd.read_csv('haberman.csv',header=None,names=header)
df[:10]
```

```python
# Check for null values
df.isnull().sum() #Great
```

```python
sb.set_style("whitegrid")
```

```python
sb.pairplot(df,hue='surv')
```

```python
df.surv.value_counts()
```

```python
df.groupby(by='op_yr')['surv'].count().plot() # People under observation from all year
```

```python
X= df.surv.values
Y=df.age.values
sb.boxplot(x=X,y=Y)
```

```python
X= df.surv.values
Y=df.aux_nodes.values
sb.boxplot(x=X,y=Y)
# People with more nodes have survived less
```
